# Noun_chunks + Logistic regression

```python
import spacy
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load spaCy's English model
nlp = spacy.load("en_core_web_sm")

# 2. Extract noun chunks
def extract_noun_chunks(text):
    doc = nlp(text)
    return " ".join([chunk.text.lower() for chunk in doc.noun_chunks])

# 3. Apply to data
user_df["noun_chunks"] = user_df["STATUS"].apply(extract_noun_chunks)

# 4. Train-test split
train_df, test_df = train_test_split(user_df, stratify=user_df["label"], test_size=0.2, random_state=42)

# 5. Vectorize noun chunks
vectorizer = CountVectorizer(ngram_range=(1,1), max_features=1000)
X_train = vectorizer.fit_transform(train_df["noun_chunks"])
X_test = vectorizer.transform(test_df["noun_chunks"])
y_train = train_df["label"]
y_test = test_df["label"]

# 6. Train classifier
clf = LogisticRegression(max_iter=1000, class_weight="balanced")
clf.fit(X_train, y_train)

# 7. Evaluate
y_pred = clf.predict(X_test)
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=["Not Neurotic", "Neurotic"]))

# 8. Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", xticklabels=["Not Neurotic", "Neurotic"],
            yticklabels=["Not Neurotic", "Neurotic"])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix (Noun Chunks)")
plt.show()
```

# PMI

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

# Extract text features using TF-IDF
vectorizer = TfidfVectorizer(ngram_range=(1, 2), stop_words='english')
X_train_text = vectorizer.fit_transform(X_train)

# Function to calculate PMI for each feature
def compute_pmi(X_train, y_train):
    n_docs = len(y_train)
    pmi_scores = []

    for i in range(X_train.shape[1]):  # Iterate through each feature
        # Count the number of documents where the feature appears (X)
        count_x = X_train[:, i].sum()

        # Count the number of documents where the feature appears and the label is 1 (Y=1)
        count_xy_1 = X_train[y_train == 1, i].sum()

        # Count the number of documents where the feature appears and the label is 0 (Y=0)
        count_xy_0 = X_train[y_train == 0, i].sum()

        # Calculate probabilities
        p_x = count_x / n_docs
        p_y_1 = sum(y_train == 1) / n_docs
        p_y_0 = sum(y_train == 0) / n_docs
        p_xy_1 = count_xy_1 / n_docs
        p_xy_0 = count_xy_0 / n_docs

        # PMI calculation
        pmi_1 = np.log(p_xy_1 / (p_x * p_y_1)) if p_xy_1 > 0 else 0
        pmi_0 = np.log(p_xy_0 / (p_x * p_y_0)) if p_xy_0 > 0 else 0

        # Average PMI for the feature
        pmi_scores.append((pmi_1 + pmi_0) / 2)

    return np.array(pmi_scores)

# Compute PMI for all features
pmi_scores = compute_pmi(X_train_text.toarray(), y_train)
print(f"PMI Scores for Features: {pmi_scores}")
```

```python
# Set the number of top features to select
top_k = 20

# Sort features by PMI score (ascending order)
top_k_features_indices = np.argsort(pmi_scores)[-top_k:]

# Get the names of the top `k` features
selected_feature_names = np.array(vectorizer.get_feature_names_out())[top_k_features_indices]

print(f"Top {top_k} Features based on PMI: {selected_feature_names}")
# Set a PMI threshold to select features
threshold = 0.1  # or -0.1 for negative PMI features

# Select features with PMI above the threshold
selected_features = [feature for i, feature in enumerate(vectorizer.get_feature_names_out()) if pmi_scores[i] > threshold]

print(f"Selected Features with PMI > {threshold}: {selected_features}")

```

```python
# Select only the relevant features for training
X_train_selected = X_train_text[:, top_k_features_indices]  # Use the indices from top_k_features_indices
X_test_selected = X_test_text[:, top_k_features_indices]

# Combine with LIWC features
X_train_combined = hstack([X_train_selected, X_train_liwc_sparse])
X_test_combined = hstack([X_test_selected, X_test_liwc_sparse])

# Train the model with the selected features
model = LogisticRegression(max_iter=1000, class_weight='balanced')
model.fit(X_train_combined, y_train)

# Predict and evaluate
y_pred = model.predict(X_test_combined)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=["Not Neurotic", "Neurotic"]))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", xticklabels=["Not Neurotic", "Neurotic"],
            yticklabels=["Not Neurotic", "Neurotic"])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix with Selected Features")
plt.show()
```

```python

# Function to get average Word2Vec embeddings for a document
def get_average_word2vec(tokens, model):
    # Get word vectors for all tokens in the document
    word_vectors = [model[token] for token in tokens if token in model]
    # If no valid vectors, return a zero vector
    if len(word_vectors) == 0:
        return np.zeros(model.vector_size)
    return np.mean(word_vectors, axis=0)

# Preprocessing function to tokenize text and get Word2Vec embeddings
def preprocess_text_word2vec(text):
    tokens = text.split()  # Simple tokenization (you can use more advanced methods if necessary)
    return get_average_word2vec(tokens, word2vec_model)

# Preprocess the data by applying Word2Vec
X_train_word2vec = X_train.apply(preprocess_text_word2vec)
X_test_word2vec = X_test.apply(preprocess_text_word2vec)
# Vectorization + Logistic Regression Pipeline

pipeline = make_pipeline(
    LogisticRegression(C=0.1, penalty='l2', solver='liblinear', max_iter=500, class_weight='balanced')
)

# Train model on resampled data
pipeline.fit(list(X_train_word2vec), y_train)


# Predict and Evaluate
y_pred = pipeline.predict(list(X_test_word2vec))
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=["Not Neurotic", "Neurotic"]))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", xticklabels=["Not Neurotic", "Neurotic"],
            yticklabels=["Not Neurotic", "Neurotic"])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```

# FastText

```python
# Convert your DataFrame into FastText training format
def prepare_fasttext_format(df, text_col, label_col, filename):
    with open(filename, 'w', encoding='utf-8') as f:
        for text, label in zip(df[text_col], df[label_col]):
            label_str = '__label__' + str(label)
            clean_text = text.replace('\n', ' ').replace('\r', ' ')
            f.write("{} {}\n".format(label_str, clean_text))

# Split user_df (mypersonality) into train/test
train_df, test_df = train_test_split(user_df, stratify=user_df["cNEU"], test_size=0.2, random_state=42)

# Convert both to FastText format
prepare_fasttext_format(train_df, "STATUS", "cNEU", "fasttext_train.txt")
prepare_fasttext_format(test_df, "STATUS", "cNEU", "fasttext_test.txt")
prepare_fasttext_format(df_test, "TEXT", "cNEU", "fasttext_eval.txt")
```

```python
import fasttext
from sklearn.metrics import classification_report, f1_score
import pandas as pd

results = []
best_model = None
best_f1_neurotic = 0
best_params = {}

# Define ranges
learning_rates = [0.01,0.1,0.2,0.3, 0.5,0.6,0.7, 1.0]
epochs = [25,40, 50,60]
ngrams = [1, 2]

# Grid search with try/except
for lr in learning_rates:
    for epoch in epochs:
        for ngram in ngrams:
            print(f"\n Training: lr={lr}, epoch={epoch}, wordNgrams={ngram}")
            try:
                model = fasttext.train_supervised(
                    'fasttext_train.txt',
                    lr=lr,
                    epoch=epoch,
                    wordNgrams=ngram,
                    verbose=2
                )

                # Predict on validation set
                y_true = test_df["cNEU"].astype(str).tolist()
                y_pred = [model.predict(text)[0][0].replace("__label__", "") for text in test_df["STATUS"]]

                # Evaluation
                report = classification_report(y_true, y_pred, target_names=["Not Neurotic", "Neurotic"], output_dict=True)
                f1_neurotic = report["Neurotic"]["f1-score"]

                results.append({
                    'lr': lr,
                    'epoch': epoch,
                    'wordNgrams': ngram,
                    'f1_neurotic': f1_neurotic,
                    'precision_neurotic': report["Neurotic"]["precision"],
                    'recall_neurotic': report["Neurotic"]["recall"],
                    'macro_f1': report["macro avg"]["f1-score"]
                })

                if f1_neurotic > best_f1_neurotic:
                    best_f1_neurotic = f1_neurotic
                    best_model = model
                    best_params = {'lr': lr, 'epoch': epoch, 'wordNgrams': ngram}

            except RuntimeError as e:
                print(f"Failed: {e} for lr={lr}, epoch={epoch}, wordNgrams={ngram}")
                continue  # skip to next combo

print("\n Best FastText parameters:", best_params)
print(f"Best F1-score: {best_f1_neurotic:.4f}")
best_model.save_model("best_fasttext_model.bin")

# Save results to DataFrame
df_results = pd.DataFrame(results).sort_values(by='f1_neurotic', ascending=False)
print("\nTop Configs:")
print(df_results.head())
```

# Focal Loss BERT OPTUNA

```python
class FocalLoss(torch.nn.Module):
    def __init__(self, alpha=None, gamma=2.0, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction

    def forward(self, inputs, targets):
        ce_loss = torch.nn.functional.cross_entropy(inputs, targets, weight=self.alpha, reduction='none')
        pt = torch.exp(-ce_loss)
        focal_loss = (1 - pt) ** self.gamma * ce_loss

        if self.reduction == 'mean':
            return focal_loss.mean()
        elif self.reduction == 'sum':
            return focal_loss.sum()
        else:
            return focal_loss
```
